```python
import requests
import pandas as pd
import re
```

```python
# FUNCION


def obtener_peli_por_genero(genero):

	lista_pelis= [] # todos lostipos de pelis (movies, short, tv series, games, etc)

	#lista_pelis_movies_short = [] # en esta lista solo salen movies y cuales shorts -> LA LISTA QUE NOS INTERESA MOSTRAR

	for año in range (1990, 2023):
		for pagina in range (1, 51):
			url = f"https://moviesdatabase.p.rapidapi.com/titles?year={año}&genre={genero}&page={pagina}&limit=50" #AÑADIDO LIMIT=50
			print(url)


		#https://moviesdatabase.p.rapidapi.com/titles?startYear=1900&genre={genero}&page=1"
		print (url)

		headers = {
		"X-RapidAPI-Key": "b5eb809b30msh73dbe0e2bd1a0b3p1b2eccjsn6ff0642c48dc",
		"X-RapidAPI-Host": "moviesdatabase.p.rapidapi.com"
		}

		response = requests.get(url, headers=headers)

		print(response.json())


		json_peli = response.json()["results"]

		if len(json_peli) > 0:
			for peli in json_peli:
				try:
					tipo_peli = peli.get('titleType', {}).get('text', '')
					if "movie" in tipo_peli.lower() or "short" in tipo_peli.lower():
						titulo = peli.get('titleText', {}).get('text', '')
						año = peli.get('releaseDate', {}).get('year', '')
						mes = peli.get('releaseDate', {}).get('month', '')
						id_peli = peli.get('id', '')
						lista_pelis.append((tipo_peli, titulo, año, mes, id_peli))
				except:
					print("Error: Ni peli ni corto? ")
					pass #cambiar x break??
		else:
			break

	
	return lista_pelis
```

COMEDIA:

```python
#Drama, Comedy, Action, Fantasy, Horror, Mystery, Romance, Thriller.

1 #Llamamos a la función


busqueda_peliculas = obtener_peli_por_genero("Comedy")

#Para ver las tuplas con cada peli:

lista_peliculas = []

for pelicula in busqueda_peliculas:
    lista_peliculas.append(pelicula)
```

```python
#2
lista_pelis_comedia = []

for index, pelicula in enumerate(busqueda_peliculas):
    lista_pelis_comedia.append(pelicula)
    print(f"Película {index + 1}: {pelicula}")
```

```python
print (lista_pelis_comedia[:3])
```

```python
#3 DATA FRAME COMEDIA
# creamos un dataframe con la lista de tuplas, y establecemos el nombre de las columnas usando el paramentro 'columns'
df = pd.DataFrame(lista_pelis_comedia, columns = ["Tipo", "Titulo", "Año", "Mes", "Id"])
df.head()
```

```python
#4 RESULTADOS CSV - COMEDIA
# # guardar los resultados en csv y en excel
df.to_csv("datos_comedia_50.csv")
```
